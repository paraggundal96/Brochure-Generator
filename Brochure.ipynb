```python
link_system_prompt = '''
You are provided with a list of links found on a webpage.
You are able to decide which of the links would be most relevant to include in a brochure about the company,
such as links to an About page, or a Company page, or Careers/Jobs pages.
You should respond in JSON as in this example:
{
    "links":[
        {"type":"about page", "url":"https://full.url/goes/here/about"},
        {"type":"careers page", "url":"https://another.full.url/careers"}
    ]
}
'''
```

```python
def user_prompt_with_links(url):

    user_prompt = '''
    Here is the list of links on the website {url} -
    Please decide which of these are relevant web links for a brochure about the company, 
    respond with the full https URL in JSON format.
    Do not include Terms of Service, Privacy, email links.

    Links (some might be relative links):
    '''
    links = fetch_website_links(url)
    user_prompt += "\n".join(links)
    return user_prompt
```

```python
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from typing import List
import os
from dotenv import load_dotenv

load_dotenv()
key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_ollama import ChatOllama


class Link(BaseModel):
    type:str = Field(description='type of link like about, careers')
    url:str = Field(description="full https URL")

class LinkSchema(BaseModel):
    links : List[Link]

ollama_model =  ChatOllama(
        model_name = "phi3:mini",
        model_provider = "ollama")

def model_(model_name):

    return ChatOpenAI(
        model = model_name,
        base_url= "https://openrouter.ai/api/v1",
        api_key= key,
        temperature= 0
    )

# Ultra Tight Prompt, no whitespaces

link_system_prompt = '''
From list of links, select relevant links to include in bochure.
Return ONLY VALID JSON no text.
{"links":[{"t":"about","u":"https://..."}]}
'''


def list_of_links(url):
    links  = fetch_website_links(url)
    return "\n".join(links)

def select_relevant_links(url, model_name):

    model = model_(model_name)
    structured_model = model.with_structured_output(LinkSchema)
    messages = [
        {"role":"system","content":link_system_prompt},
        {"role":"user","content":list_of_links(url)}
    ]

    response = structured_model.invoke(messages)
    return response.links

models = [
    ollama_model,
    "mistralai/mistral-7b-instruct",
    "openai/gpt-oss-120b:free",
    "google/gemma-3-12b-it:free",
    "qwen/qwen3-coder:free",
    "nvidia/nemotron-3-super-120b-a12b:free"
]

for m in models:
    try:
        result = select_relevant_links("https://github.com/paraggundal96", m)
        print(result)
    except Exception as e:
        print(m, "FAILED:", e)
```
